```python
import sys
import os
from dotenv import load_dotenv, find_dotenv
import geopandas as gpd
import matplotlib.pyplot as plt
import fiona
from shapely.geometry import box
import pandas as pd
import contextily as ctx

load_dotenv(find_dotenv())

DATA_PATH = os.getenv("DATA_PATH")
```

```python
def load_gdb(filepath, crs=3857):
    layers = fiona.listlayers(filepath)
    layer = layers[0]
    gdf = gpd.read_file(filepath, layer=layer)

    # convert crs
    if gdf.crs and gdf.crs.to_epsg() != crs:
        gdf = gdf.to_crs(epsg=crs)
    return gdf
```

```python
INPUT_FILENAME = "EPIC-LA_Case_History_view.gdb"
INPUT_FILEPATH = os.path.join(DATA_PATH, "raw_data", INPUT_FILENAME)
```

```python
gdf = load_gdb(INPUT_FILEPATH)
```
